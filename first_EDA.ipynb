```python
# Настройка стилей
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 100)

# Путь к данным
DATA_PATH = '/content/drive/MyDrive/ieee-db/'
transaction_file = DATA_PATH + 'train_transaction.csv'
identity_file = DATA_PATH + 'train_identity.csv'

print("📁 Загружаем фрод-транзакции и соединяем с identity...\n")

# 1. Загружаем identity таблицу полностью (она меньше размером)
print("Загрузка identity таблицы...")
identity_df = pd.read_csv(identity_file)
print(f" Identity таблица: {identity_df.shape[0]} строк, {identity_df.shape[1]} столбцов")

# 2. Читаем transaction чанками и собираем только фрод
chunk_size = 50000  # увеличим размер чанка для скорости
fraud_chunks = []
total_fraud = 0
max_fraud_to_collect = 500  # сколько фрода хотим собрать

print("\n🔍 Поиск фрод-транзакций в transaction таблице...")

for i, chunk in enumerate(pd.read_csv(transaction_file, chunksize=chunk_size)):
    fraud_chunk = chunk[chunk['isFraud'] == 1]
    fraud_count = len(fraud_chunk)
    
    if fraud_count > 0:
        fraud_chunks.append(fraud_chunk)
        total_fraud += fraud_count
        print(f"  Чанк {i+1}: найдено {fraud_count} фрод-строк (всего: {total_fraud})")
    
    # Останавливаемся, когда наберем достаточно фрода
    if total_fraud >= max_fraud_to_collect:
        break

# 3. Объединяем все собранные фрод-чанки
if fraud_chunks:
    fraud_df = pd.concat(fraud_chunks, ignore_index=True)
    print(f"\n Всего собрано {len(fraud_df)} мошеннических транзакций")
    
    # 4. Соединяем с identity таблицей
    print("\n🔗 Соединяем с identity информацией...")
    fraud_with_identity = fraud_df.merge(identity_df, on='TransactionID', how='left')
    
    print(f" После соединения: {fraud_with_identity.shape[0]} строк, {fraud_with_identity.shape[1]} столбцов")
    
    # 5. Сколько фрода имеет identity информацию
    has_identity = fraud_with_identity['DeviceType'].notna().sum()
    print(f"\n Из {len(fraud_with_identity)} фрод-транзакций:")
    print(f"   - С identity информацией: {has_identity} ({has_identity/len(fraud_with_identity)*100:.1f}%)")
    print(f"   - Без identity: {len(fraud_with_identity) - has_identity}")
    
    # 6. Показываем результаты
    print("\n📋 ПЕРВЫЕ 20 ФРОД-ТРАНЗАКЦИЙ С IDENTITY ИНФОРМАЦИЕЙ:")
    
    # Выбираем ключевые колонки для просмотра
    transaction_cols = ['TransactionID', 'isFraud', 'TransactionAmt', 'ProductCD', 'card1', 'card2', 'card3', 'card4', 'card5', 'card6']
    identity_cols = ['DeviceType', 'DeviceInfo', 'id_01', 'id_02', 'id_03', 'id_04', 'id_05', 
                     'id_12', 'id_13', 'id_14', 'id_15', 'id_30', 'id_31', 'id_32', 'id_33', 'id_34', 'id_35', 'id_36', 'id_37', 'id_38']
    
    # Берем только существующие колонки
    доступно_cols = [col for col in transaction_cols + identity_cols if col in fraud_with_identity.columns]
    
    display(fraud_with_identity[доступно_cols].head(20))
    
    # 7. Дополнительный анализ: сравнение фрода с identity и без
    print("\n СРАВНЕНИЕ ФРОДА С IDENTITY И БЕЗ:")
    
    fraud_with_id = fraud_with_identity[fraud_with_identity['DeviceType'].notna()]
    fraud_without_id = fraud_with_identity[fraud_with_identity['DeviceType'].isna()]
    
    if len(fraud_with_id) > 0 and len(fraud_without_id) > 0:
        print(f"\nСредняя сумма транзакции:")
        print(f"   - С identity: ${fraud_with_id['TransactionAmt'].mean():.2f}")
        print(f"   - Без identity: ${fraud_without_id['TransactionAmt'].mean():.2f}")
    
    # 8. Визуализация
    fig, axes = plt.subplots(1, 3, figsize=(15, 4))
    
    # График 1: Типы устройств при фроде
    if 'DeviceType' in fraud_with_identity.columns:
        device_counts = fraud_with_identity['DeviceType'].value_counts()
        axes[0].pie(device_counts.values, labels=device_counts.index, autopct='%1.1f%%')
        axes[0].set_title('Типы устройств при фроде')
    
    # График 2: Операционные системы
    if 'id_30' in fraud_with_identity.columns:
        os_counts = fraud_with_identity['id_30'].value_counts().head(5)
        axes[1].bar(range(len(os_counts)), os_counts.values)
        axes[1].set_xticks(range(len(os_counts)))
        axes[1].set_xticklabels(os_counts.index, rotation=45)
        axes[1].set_title('Топ-5 ОС при фроде')
    
    # График 3: Распределение сумм
    axes[2].hist(fraud_with_identity['TransactionAmt'], bins=30, color='red', alpha=0.7)
    axes[2].set_xlabel('Сумма транзакции')
    axes[2].set_ylabel('Частота')
    axes[2].set_title('Распределение сумм фрода')
    
    plt.tight_layout()
    plt.show()
    
else:
    print(" Фрод-транзакции не найдены")
```
